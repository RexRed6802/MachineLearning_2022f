```python
X = data.iloc[:,0:7]
y = data.iloc[:,7:]
```

```python
X.head()
```

```python
y.value_counts()
```

```python
from sklearn import preprocessing
zscore = preprocessing.StandardScaler()
X = zscore.fit_transform(X)
```

```python
y = np.array(y)
```

```python
def toCategory(lst):
  t = []
  for i in range(0,len(lst)):
      k = lst
      if(k[i] == 'A'):
          t.append(0)
      elif(k[i] == 'B'):
          t.append(1)
      elif(k[i] == 'C'):
         t.append(2)
  t = to_categorical(t, 3)
  return t;
y = toCategory(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state = 42)
```

```python
#from sklearn.model_selection import train_test_split
#X_train, X_val, y_train, y_val = train_test_split(X, t, test_size=0.33, random_state=42)
```

```python
model = Sequential()
model.add(Dense(50, input_dim=7, activation='relu'))
model.add(Dense(35, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', 
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=33, epochs=20, validation_data=(X_test, y_test))
```

```python
model.fit(X,y,batch_size=33, epochs=20,)
```

```python
zscore2 = preprocessing.StandardScaler()
test_data = zscore2.fit_transform(test_data)
```

```python
answer = np.argmax(model.predict(test_data),axis = -1)
```

```python
answer
```

```python
ans = []
for a in answer:
  if(a == 0):
    ans.append("A")
  elif(a == 1):
    ans.append("B")
  elif(a == 2):
    ans.append("C")
```

```python
id = np.linspace(1,1200,1200,dtype = "int")
a = np.concatenate((id,ans))
a = a.reshape(2,1200)
a = np.transpose(a, axes=None)
```
